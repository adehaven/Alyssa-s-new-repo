```python
import requests
from bs4 import BeautifulSoup
import csv

url = "https://archiveofourown.org/works?commit=Sort+and+Filter&work_search%5Bsort_column%5D=kudos_count&work_search%5Bother_tag_names%5D=Canon+Universe&work_search%5Bexcluded_tag_names%5D=&work_search%5Bcrossover%5D=&work_search%5Bcomplete%5D=&work_search%5Bwords_from%5D=&work_search%5Bwords_to%5D=&work_search%5Bdate_from%5D=&work_search%5Bdate_to%5D=&work_search%5Bquery%5D=&work_search%5Blanguage_id%5D=en&tag_id=Harry+Potter+-+J*d*+K*d*+Rowling"

response = requests.get(url)
response.raise_for_status()  


soup = BeautifulSoup(response.text, 'html.parser')


entries = soup.find_all('li', class_='work')


data = []


for entry in entries[:10]:
    
    def safe_get_text(soup_obj, class_name):
        element = soup_obj.find(class_=class_name)
        return element.get_text(strip=True) if element else 'N/A'

    title = safe_get_text(entry, 'heading')
    author = safe_get_text(entry, 'author')
    kudos = safe_get_text(entry, 'kudos')
    comments = safe_get_text(entry, 'comments')
    hits = safe_get_text(entry, 'hits')
    url = entry.find('a', class_='heading')['href'] if entry.find('a', class_='heading') else 'N/A'
    summary = safe_get_text(entry, 'summary')  

    
    data.append([title, author, kudos, comments, hits, summary, url])


with open('ao3_stories_with_summaries.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Author', 'Kudos', 'Comments', 'Hits', 'Summary', 'URL'])
    writer.writerows(data)

print("CSV file 'ao3--stories.csv' created successfully.")

```

```python
import pandas as pd
from collections import Counter


characters = [
    "Hermione Granger", "Draco Malfoy", "Dolores Umbridge", "Tom Riddle | Voldemort",
    "Astoria Greengrass", "Graham Montague", "Severus Snape", "Moody", "Harry Potter",
    "Ron Weasley", "Ginny Weasley", "Tonks", "Remus Lupin", "Bellatrix Lestrange",
    "Dolohov", "Rosier", "Lucius Malfoy", "Pansy Parkinson", "Padma Patil", "Lavender Brown"
]


def most_common_character(df):
    try:
        summaries = df['summary'].dropna().astype(str).str.cat(sep=' ')
    except KeyError:
        return ("None", 0)
    
    word_counts = Counter()
    for character in characters:
        word_counts[character] = summaries.count(character)
    
    most_common = word_counts.most_common(1)
    return most_common[0] if most_common else ("None", 0)


df1 = pd.read_csv('harry_potter_fanfiction.csv')
df2 = pd.read_csv('Hermione-Fanfiction.csv')
df3 = pd.read_csv('ao3_stories_with_summaries.csv')


most_common_df1 = most_common_character(df1)
most_common_df2 = most_common_character(df2)
most_common_df3 = most_common_character(df3)


print("Most common character in harry_potter_fanfiction.csv:", most_common_df1)
print("Most common character in hermione_draco_fanfiction.csv:", most_common_df2)
print("Most common character in ao3_stories_with_summaries.csv:", most_common_df3)

```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score
import nltk
from nltk.corpus import stopwords


nltk.download('stopwords')


df_ao3_stories = pd.read_csv('ao3_stories_with_summaries.csv')
df_harry_potter = pd.read_csv('harry_potter_fanfiction.csv')
df_hermione_fanfiction = pd.read_csv('Hermione-Fanfiction.csv')


def most_common_character(df, column_name='character'):
    if column_name in df.columns:
        return df[column_name].mode()[0]  
    else:
        return "Character column not found."


print("Most common character in ao3_stories_with_summaries.csv:", most_common_character(df_ao3_stories))
print("Most common character in harry_potter_fanfiction.csv:", most_common_character(df_harry_potter))
print("Most common character in Hermione-Fanfiction.csv:", most_common_character(df_hermione_fanfiction))


df_combined = pd.concat([df_harry_potter, df_hermione_fanfiction], ignore_index=True)


stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    return ' '.join([word.lower() for word in text.split() if word.lower() not in stop_words])


if 'summary' in df_ao3_stories.columns:
    df_ao3_stories['summary'] = df_ao3_stories['summary'].apply(preprocess_text)
if 'summary' in df_harry_potter.columns:
    df_harry_potter['summary'] = df_harry_potter['summary'].apply(preprocess_text)
if 'summary' in df_hermione_fanfiction.columns:
    df_hermione_fanfiction['summary'] = df_hermione_fanfiction['summary'].apply(preprocess_text)
if 'summary' in df_combined.columns:
    df_combined['summary'] = df_combined['summary'].apply(preprocess_text)

data = {
    'summary': [
        'A thrilling adventure where Harry Potter faces new challenges.',
        'Romantic story between Hermione and Draco with lots of drama.',
        'A dark tale featuring Voldemort and his followers.',
        'A light-hearted story about Ron Weasley and his misadventures.',
        'Hermione Granger and Severus Snape team up in a mystery novel.'
    ],
    'genre': ['Adventure', 'Romance', 'Dark', 'Comedy', 'Mystery']
}

df_labeled = pd.DataFrame(data)


X_train, X_test, y_train, y_test = train_test_split(df_labeled['summary'], df_labeled['genre'], test_size=0.2, random_state=42)


model = make_pipeline(TfidfVectorizer(), MultinomialNB())


model.fit(X_train, y_train)


y_pred = model.predict(X_test)


print(f"Accuracy: {accuracy_score(y_test, y_pred)}")


def predict_genre(summary):
    return model.predict([summary])[0]


if 'summary' in df_combined.columns:
    df_combined['predicted_genre'] = df_combined['summary'].apply(predict_genre)


print(df_combined[['summary', 'predicted_genre']])
```

```python
import pandas as pd
import nltk
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from collections import Counter

nltk.download('stopwords')

negative_words = set([
    'bad', 'worse', 'worst', 'awful', 'terrible', 'horrible', 'hate', 'angry',
    'sad', 'pain', 'fear', 'ugly', 'disgust', 'horrific', 'negative', 'bleed', 'broken', 'hurt', 'apart','seperate','died','death','dying'
])

stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    words = text.lower().split()
    filtered_words = [word for word in words if word not in stop_words]
    return filtered_words

def extract_text_from_html(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')
        text = soup.get_text()
    return text

def count_negative_words(text, negative_words):
    words = preprocess_text(text)
    negative_count = Counter(word for word in words if word in negative_words)
    return negative_count

story1_path = 'All_the_Young_Dudes.html'
story2_path = 'Draco_Malfoy_and_the.html'

story1_text = extract_text_from_html(story1_path)
story2_text = extract_text_from_html(story2_path)

story1_negative_counts = count_negative_words(story1_text, negative_words)
story2_negative_counts = count_negative_words(story2_text, negative_words)

print("Negative words in All_the_Young_Dudes:")
print(story1_negative_counts)

print("\nNegative words in Draco_Malfoy_and_the:")
print(story2_negative_counts)

comparison = {
    'word': [],
    'Canon': [],
    'Non-canon': []
}
for word in negative_words:
    story1_count = story1_negative_counts.get(word, 0)
    story2_count = story2_negative_counts.get(word, 0)
    if story1_count > 0 or story2_count > 0:
        comparison['word'].append(word)
        comparison['Canon'].append(story1_count)
        comparison['Non-canon'].append(story2_count)

comparison_df = pd.DataFrame(comparison)

print("\nComparison of negative words between the two stories:")
print(comparison_df)
```

```python
import pandas as pd
import nltk
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from collections import Counter
import matplotlib.pyplot as plt

nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

negative_words = set([
    'bad', 'worse', 'worst', 'awful', 'terrible', 'horrible', 'hate', 'angry', 'sad', 'depressing', 'excruciating', 'crucio','blood','death','dying','dark','curse'
])

positive_words = set([
    'good', 'better', 'best', 'wonderful', 'excellent', 'amazing', 'love', 'happy', 'joyful', 'positive','live','beautiful','close','light','heal','spell'
])

def preprocess_text(text):
    return [word.lower() for word in text.split() if word.lower() not in stop_words]

# Function to extract text from HTML file
def extract_text_from_html(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')
        return soup.get_text()

def count_words(text, words):
    word_list = preprocess_text(text)
    word_counter = Counter([word for word in word_list if word in words])
    return word_counter

file_path = 'Manacled.html'
story_text = extract_text_from_html(file_path)

negative_word_counts = count_words(story_text, negative_words)
positive_word_counts = count_words(story_text, positive_words)

df_negative = pd.DataFrame(negative_word_counts.items(), columns=['Word', 'Count']).sort_values(by='Count', ascending=False)
df_positive = pd.DataFrame(positive_word_counts.items(), columns=['Word', 'Count']).sort_values(by='Count', ascending=False)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.bar(df_negative['Word'], df_negative['Count'], color='gray')
plt.xlabel('Negative Words')
plt.ylabel('Frequency')
plt.title('Most Frequent Negative Words')

plt.subplot(1, 2, 2)
plt.bar(df_positive['Word'], df_positive['Count'], color='pink')
plt.xlabel('Positive Words')
plt.ylabel('Frequency')
plt.title('Most Frequent Positive Words')

plt.tight_layout()
plt.show()
```

Manacled's network was especially important to me, as I wanted to determine if the character interactions and network had any relevance to the theme. Hermione Granger remains the main character, which is in accordance with the findings of the original study that identified her as a primary source for non-canonical information. Applying *Networkx*, assess every character interaction and count them to help me graph.The examining of these central characters might help discover a pattern where those with higher centrality often exhibit traits such as moral ambiguity or emotional turmoil. This could help us understand the relationship between narrative importance and the complexity of character development.

```python
import re
from bs4 import BeautifulSoup
import networkx as nx
from collections import Counter

with open('Manacled.html', 'r', encoding='utf-8') as file:
    content = file.read()

soup = BeautifulSoup(content, 'html.parser')
text = soup.get_text()

characters = [
    "Hermione Granger", "Draco Malfoy", "Dolores Umbridge", "Tom Riddle", "Voldemort",
    "Astoria Greengrass", "Graham Montague", "Severus Snape", "Moody",
    "Harry Potter", "Ron Weasley", "Ginny Weasley", "Tonks", "Remus Lupin",
    "Bellatrix Lestrange", "Dolohov", "Rosier", "Lucius Malfoy", "Pansy Parkinson",
    "Padma Patil", "Lavender Brown"
]

char_pattern = re.compile(r'\b(?:' + '|'.join(re.escape(char) for char in characters) + r')\b')

char_occurrences = char_pattern.findall(text)

interaction_pairs = [(char_occurrences[i], char_occurrences[i+1]) for i in range(len(char_occurrences)-1) if char_occurrences[i] != char_occurrences[i+1]]

interaction_counts = Counter(interaction_pairs)
```

```python
G = nx.Graph()

for (char1, char2), weight in interaction_counts.items():
    if G.has_edge(char1, char2):
        G[char1][char2]['weight'] += weight
    else:
        G.add_edge(char1, char2, weight=weight)

degree_centrality = nx.degree_centrality(G)

for character, centrality in degree_centrality.items():
    print(f"{character}: {centrality:.4f}")
```

Our famous villain exhibits the highest degree of centrality possible. Consequently, dark themes emerge in the story. 


```python
import re
import networkx as nx
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
from collections import Counter

with open('Manacled.html', 'r', encoding='utf-8') as file:
    content = file.read()

soup = BeautifulSoup(content, 'html.parser')
text = soup.get_text()

characters = [
    "Hermione Granger", "Draco Malfoy", "Dolores Umbridge", "Tom Riddle", "Voldemort",
    "Astoria Greengrass", "Graham Montague", "Severus Snape", "Moody",
    "Harry Potter", "Ron Weasley", "Ginny Weasley", "Tonks", "Remus Lupin",
    "Bellatrix Lestrange", "Dolohov", "Rosier", "Lucius Malfoy", "Pansy Parkinson",
    "Padma Patil", "Lavender Brown"
]

char_pattern = re.compile(r'\b(?:' + '|'.join(re.escape(char) for char in characters) + r')\b')

char_occurrences = char_pattern.findall(text)

interaction_pairs = [(char_occurrences[i], char_occurrences[i+1]) for i in range(len(char_occurrences)-1) if char_occurrences[i] != char_occurrences[i+1]]

interaction_counts = Counter(interaction_pairs)

G = nx.Graph()

for (char1, char2), weight in interaction_counts.items():
    if G.has_edge(char1, char2):
        G[char1][char2]['weight'] += weight
    else:
        G.add_edge(char1, char2, weight=weight)

degree_centrality = nx.degree_centrality(G)

for character, centrality in degree_centrality.items():
    print(f"{character}: {centrality:.4f}")

plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.3)  
nx.draw_networkx_nodes(G, pos, node_size=300, node_color='orange')
nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.7)
nx.draw_networkx_labels(G, pos, font_size=12, font_family='sans-serif')

edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='pink')

plt.title('Degree Network for "Manacled"')
plt.axis('off')  
plt.show()
```

you can see the most central character with the most influence, in the middle.
